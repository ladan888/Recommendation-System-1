### **2.4.5|** Control of Missing and Duplicate values

Earlier we merged the two datasets into a single data frame. We want to check and make sure the data merged properly and that no null or missing values were created in the process.


````python
# We will use 'data.isnull()' to generate the boolean values for whether or not the dataset has null values
# Then we will use '.sum()' take the sum of any null values in the dataset for each variable
# Then we will take the sum again to get an overall number for the dataset
data.isnull().sum().sum()
````

Since the value above was equal to zero, we now know that our dataset has no null values. 

Next, we want to do an analysis of duplicate values in our dataset. We will create a dictionary of values showing the duplicated values in each category of data. We can then call on this if needed in later steps to see the unique values in each category.

````python
# Creating an empty dictionary and empty array to store the unique values in 
duplicates = {}
# creating a dictionary with all of the unique values for each variable
for column in data:                 # iterating through each variable
  dups=list(data[column].unique())  # generating a list of all unique values in each variable category
  duplicates[column] = dups         # storing these unique values to a dictionary
  print(column,': ',dups)           # printing the results 
````

### **2.4.6|** Frequency Counts

We want to get a better understanding of the duplicate value counts we found above, specifically for our categorical datatypes. We will generate dictionaries showing the frequencies of each repeated value in a category.

````python
# Finding the frequency counts for categorical data
count = {}

for column in data:                     # iteating through columns of the dataset
  if data[column].dtype == 'object':    # checking to see if the column's datatype is an object
    for item in list(data[column]):     # going through row in the column
      if item in count:                 # checking to see if the item has be counted already
        count[item] += 1                # if the item already exists in the dictionary, then we add 1 to its count value
      else:                   
        count[item] = 1                 # the item doesn't exist so we subtract 1
    print(column,'frequencies: \n', count,'\n')
    count={}                            # resetting the counter to check the next column
      
````

When it comes to building earthquake safe buildings, both height and area design are a large factor into a building’s overall safety. Engineers often refer to a design spectrum similar to the one shown below to understand the relationship between a buildings height and how much shaking it will experience during an earthquake [1].



earthquake_tall_buildings_fig1.gif
> **Figure 14: Earthquake Safe Building Design Spectrum**

In the image above, the solid black line explains the relationship between T and S(g). These variables are explained below:
* ***T:*** the building's fundamental period, or the rate at which it will move back and forth in response to the ground shaking
* ***S(g):*** the spectral acceleration, or the shaking force a building is expected to experience at a given site

The height of the building is the main determining of the fundamental period [2]. The dark line in this chart is higher in the lowrise section of the chart, and lowest in the skyscrapers section. This means that shorter buildings will experience more aggressive shaking forces in the event of an earthquake than taller buildings will, resulting in more damage.
